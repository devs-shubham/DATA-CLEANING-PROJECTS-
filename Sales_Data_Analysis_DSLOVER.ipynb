 Q.2) Which one was the Most Selling Product
 .By Quantity


 .By Revenue



```python
df.head()
most_Quantity = df.groupby("Product")["Quantity"].sum().sort_values(ascending=False)
most_Quantity
```

```python
type(most_Quantity)
```

```python
most_Quantity = most_Quantity.reset_index()
most_Quantity
```

```python
type(most_Quantity)
```

```python
# Bar graph
plt.figure(figsize=(8,3))
plt.bar(most_Quantity["Product"],most_Quantity["Quantity"] , color = ["red", 'black', 'green', "yellow" , "cyan"] , width=0.6)
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.title("Most Selling Product")
plt.show()
```

# Revenue


```python
df["Revenue"] = df["Quantity"] * df["Price"]
df.head()
```

```python

most_revenue = df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
most_revenue
```

```python
# Bar graph
plt.figure(figsize=(8,3))
plt.bar(most_revenue.index, most_revenue.values , color = ["red", 'black', 'green', "yellow" , "cyan"] , width=0.6)
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.title("Most Selling Product by Revenue")
plt.show()
```

Q.9) Was the revenue increasing or decreasing over the time?

```python
df.head()
Revenue_over_time = df.groupby("Date")["Revenue"].sum()
Revenue_over_time
```

```python
# By visulaization methods
plt.figure(figsize=(10,5))
plt.plot(Revenue_over_time);
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Revenue Over Time")
plt.show()
```

Q.10) What was the Average 'Quantity Sold' & 'Average Revenue' for each product ?

```python
df.groupby("Product")[["Quantity", "Revenue"]].agg({
    "Quantity": "mean",
    "Revenue": "mean"
})
```
